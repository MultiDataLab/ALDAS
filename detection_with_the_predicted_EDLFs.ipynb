```python
import numpy as np
from sklearn.model_selection import StratifiedKFold
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Dropout, Flatten, Dense
from sklearn.metrics import accuracy_score

# Define the CNN model
def get_model(input_shape):
    model = Sequential()
    model.add(Conv1D(32, kernel_size=3, activation='relu', input_shape=input_shape))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Conv1D(64, kernel_size=3, activation='relu'))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Flatten())
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation='sigmoid'))
    return model

# Define the input shape
input_shape = (3, 1)  # Assuming 3 features per sample

# Define k-fold cross-validation
skf = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)

# Initialize lists to store train and validation scores
train_scores = []
val_scores = []

# Loop through each fold
for train_index, val_index in skf.split(X_train_soft, y_train):
    # Split data into train and validation sets
    X_train_fold, X_val_fold = X_train_soft[train_index], X_train_soft[val_index]
    y_train_fold, y_val_fold = y_train[train_index], y_train[val_index]

    # Reshape input data to match input_shape
    X_train_fold = np.reshape(X_train_fold, (X_train_fold.shape[0], X_train_fold.shape[1], 1))
    X_val_fold = np.reshape(X_val_fold, (X_val_fold.shape[0], X_val_fold.shape[1], 1))

    # Create and compile the model
    model = get_model(input_shape)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    # Train the model
    model.fit(X_train_fold, y_train_fold, epochs=10, batch_size=32, verbose=0)

    # Evaluate the model on train and validation data
    train_score = model.evaluate(X_train_fold, y_train_fold, verbose=0)[1]
    val_score = model.evaluate(X_val_fold, y_val_fold, verbose=0)[1]

    # Append scores to lists
    train_scores.append(train_score)
    val_scores.append(val_score)

# Calculate mean train and validation scores
mean_train_score = np.mean(train_scores)
mean_val_score = np.mean(val_scores)

# Print the mean scores
print("Mean train accuracy:", mean_train_score)
print("Mean validation accuracy:", mean_val_score)
```
